```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# بارگذاری فایل CSV
df1= pd.read_csv(r"C:\Users\Lenovo\Desktop\dr.hoseyni\separated\R-ARG.csv",low_memory=False)
df2= pd.read_csv(r"C:\Users\Lenovo\Desktop\dr.hoseyni\separated\R-ASN.csv",low_memory=False)
df3= pd.read_csv(r"C:\Users\Lenovo\Desktop\dr.hoseyni\separated\R-ASP.csv",low_memory=False)
df4= pd.read_csv(r"C:\Users\Lenovo\Desktop\dr.hoseyni\separated\R-GLN.csv",low_memory=False)
df5= pd.read_csv(r"C:\Users\Lenovo\Desktop\dr.hoseyni\separated\R-GLU.csv",low_memory=False)
df6= pd.read_csv(r"C:\Users\Lenovo\Desktop\dr.hoseyni\separated\R-LEU.csv",low_memory=False)
df7= pd.read_csv(r"C:\Users\Lenovo\Desktop\dr.hoseyni\separated\R-LYS.csv",low_memory=False)
df8= pd.read_csv(r"C:\Users\Lenovo\Desktop\dr.hoseyni\separated\R-THR.csv",low_memory=False)


df_sorted1 = df1.sort_values(by='TOTAL', ascending=True)
df_sorted2 = df2.sort_values(by='TOTAL', ascending=True)
df_sorted3 = df3.sort_values(by='TOTAL', ascending=True)
df_sorted4 = df4.sort_values(by='TOTAL', ascending=True)
df_sorted5 = df5.sort_values(by='TOTAL', ascending=True)
df_sorted6 = df6.sort_values(by='TOTAL', ascending=True)
df_sorted7 = df7.sort_values(by='TOTAL', ascending=True)
df_sorted8 = df8.sort_values(by='TOTAL', ascending=True)


df_sorted1.to_csv('Neg_TOTAL_versus_ARG.csv', index=False)
df_sorted2.to_csv('Neg_TOTAL_versus_ASN.csv', index=False)
df_sorted3.to_csv('Neg_TOTAL_versus_ASP.csv', index=False)
df_sorted4.to_csv('Neg_TOTAL_versus_GLN.csv', index=False)
df_sorted5.to_csv('Neg_TOTAL_versus_GLU.csv', index=False)
df_sorted6.to_csv('Neg_TOTAL_versus_LEU.csv', index=False)
df_sorted7.to_csv('Neg_TOTAL_versus_LYS.csv', index=False)
df_sorted8.to_csv('Neg_TOTAL_versus_THR.csv', index=False)


#print(df_sorted[['Residue','Residue Number','TOTAL']].head(10))
```

```python
x = df_sorted1 ['TOTAL']  # جایگزین با نام ستون X
y = df_sorted1 ['Residue Number']  # جایگزین با نام ستون Y
categories = df_sorted1 ['Residue']  # نام ستون دسته‌بندی
plt.figure(figsize=(8,6))
# تعیین رنگ‌ها بر اساس دسته‌بندی
unique_categories = categories.unique()
colors = plt.cm.viridis(np.linspace(0, 1, len(unique_categories)))

#plt.xticks(np.arange(-300, 10, 10))  # خطوط در محور X با فاصله 0.5
#plt.yticks(np.arange(600, 710, ))  # خطوط در محور Y با فاصله 0.1

# ایجاد دیکشنری برای نگاشت رنگ به دسته‌ها
category_color_map = {cat: colors[i] for i, cat in enumerate(unique_categories)}

# رسم نقاط با رنگ‌های مختلف
for cat in unique_categories:
    subset = df_sorted1 [df_sorted1 ['Residue'] == cat]
    plt.scatter(subset['TOTAL'], subset['Residue Number'], color=category_color_map[cat], label=cat,s=5)

plt.title('Scatter Plot of TOTAL versus ARG')
plt.xlabel('TOTAL')
plt.ylabel('Residue Number')
plt.legend(title='Residue')
plt.grid()
plt.savefig('TOTAL_versus_Residue_Number_ARG.png',dpi=300)
plt.show()
```
